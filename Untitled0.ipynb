```python

plt.figure(figsize=(12, 6))


def plot_count(feature):
    plt.figure(figsize=(12, 6))
    sns.countplot(x=feature, hue='loan_status', data=df, palette='viridis')
    plt.title(f'Loan Status Distribution with respect to {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.legend(title='Loan Status')
    plt.xticks(rotation=45)
    plt.show()

categorical_features = ['loanamount', 'loan_amount_term', 'applicantincome']


for feature in categorical_features:
    plot_count(feature)
```
